```python
import numpy as np
import gym

actions = {
    'Left': 0,
    'Down': 1,
    'Right': 2, 
    'Up': 3
}
env = gym.make("FrozenLake-v0", is_slippery=True)
alpha=0.01
gamma=0.95
Q=np.zeros([env.observation_space.n,env.action_space.n])
for episode in range(10000):
    s=env.reset()
    for i in range(100):
        a=np.argmax(Q[s,:])
        if np.random.rand()<0.2:
            a=env.action_space.sample()
        s_,r,done,_=env.step(0)
        if done and r==0:
            r=-1.0
        Q[s,a]=(1-alpha)*Q[s,a]+alpha*(r+gamma*np.max(Q[s_,:]))
        s=s_
        if done:
            break
            
for i in range(100):
    s=env.reset()
    a=np.argmax(Q[s,:])
    s_,r,done,_=env.step(0)
    env.render()
    if done:
        break
```

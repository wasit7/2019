```python
%matplotlib inline
import numpy as np
import matplotlib.cm as cm
from matplotlib import pyplot as plt
plt.style.use('ggplot')
```

```python
x=[]
yt=[]
for i in range(100):
    x.append( [0.5+np.random.rand(), 0.5+np.random.rand()])
    x.append( [-0.5+np.random.rand(), -0.5+np.random.rand()])
    yt.append([0 , 1])
    yt.append([0 , 1])

for i in range(100):
    x.append( [0.5+np.random.rand(), -0.5+np.random.rand()])
    x.append( [-0.5+np.random.rand(), 0.5+np.random.rand()])
    yt.append([1 , 0])
    yt.append([1 , 0])
    
x=np.array(x)
yt=np.array(yt)

plt.plot( x[yt[:,1]==1,0], x[yt[:,1]==1,1], 'ob')
plt.plot( x[yt[:,0]==1,0], x[yt[:,0]==1,1], 'or')
```

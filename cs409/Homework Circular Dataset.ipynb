```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
```

```python
n=200
theta=180*(np.random.random(n))
r=0.0+1.0*(np.random.randn(n))
x11=r*np.cos(theta)
x12=r*np.sin(theta)

yt=np.ones(2*n)
yt[:n]=-1

theta=180*(np.random.random(n))
r=5.0+1.0*(np.random.randn(n))
x21=r*np.cos(theta)
x22=r*np.sin(theta)

x1=np.concatenate((x11,x21),axis=None)
x2=np.concatenate((x12,x22),axis=None)
plt.plot(x1[yt==-1],x2[yt==-1],'or')
plt.plot(x1[yt==1],x2[yt==1],'ob')
```

```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
plt.style.use('ggplot')
```

```python
n=200
def normal(n):
    return np.random.randn(n)
x1=np.concatenate((normal(n)-2,normal(n)+1),axis=None)
x2=np.concatenate((normal(n)-2,normal(n)+2),axis=None)
yt=np.ones(2*n)
yt[:n]=-1
plt.plot(x1[yt==-1],x2[yt==-1],'or')
plt.plot(x1[yt==1],x2[yt==1],'ob')
```

```python
import numpy as np
alpha=0.01
theta1=0.0
theta2=0.0
theta3=0.0
n=100
log=np.zeros((n,4)) #[loss,theta1,theta2,theta3]
def forward(x1,x2):
    z = theta1*x1 + theta2*x2 + theta3
    return np.tanh(z)
for i in range(100):
    yp=forward(x1,x2)
    theta1=theta1-alpha*np.average((yp-yt)*(1-yp*yp)*x1)
    theta2=theta2-alpha*np.average((yp-yt)*(1-yp*yp)*x2)
    theta3=theta3-alpha*np.average((yp-yt)*(1-yp*yp))
    loss=0.5*np.average((yp-yt)*(yp-yt))
    log[i,:]=(loss,theta1,theta2,theta3)
    if i%(n//10)==0:
        print(loss)
        
fig, ax = plt.subplots()
ax.plot(log[:,0], 'r', label='Loss')
ax.plot(log[:,1], 'g--', label='theta1')
ax.plot(log[:,2], 'b--', label='theta2')
ax.plot(log[:,3], 'm', label='theta3')
legend = ax.legend(loc='upper right', shadow=True, fontsize='small')
```

```python
step = 0.5
mx2, mx1 = np.mgrid[slice(-6,6, step),
                    slice(-6,6, step)]
cmap = plt.get_cmap('YlGnBu_r')
plt.pcolormesh(mx1, mx2, forward(mx1,mx2), cmap=cmap)

plt.plot(x1[yt==-1],x2[yt==-1],'or')
plt.plot(x1[yt==1],x2[yt==1],'ob')
```
